```python
import numpy as np
import matplotlib as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

# Step 2
### train model on this data and keep checking accuracy; see how well it classifies 


## KNN

```python
label_encoder_pos = LabelEncoder()
label_encoder_league = LabelEncoder()
prospects2023['Pos_encoded'] = label_encoder_pos.fit_transform(prospects2023['Pos'])
prospects2023['League_encoded'] = label_encoder_league.fit_transform(prospects2023['League'])
```

```python
prospects2023['NBA-EWA'] = prospects2023['NBA-EWA'].replace('-', 0)
prospects2023['NBA-EWA'] = pd.to_numeric(prospects2023['NBA-EWA'])
```

```python
y = prospects2023[["Need"]]
x = prospects2023[['Pos_encoded', 'Age', 'Pts', 'Reb', 'Ast', 'PER', 'League_encoded']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Standardize the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train_scaled, y_train)

# Predict and evaluate
y_pred = knn.predict(x_test_scaled)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
